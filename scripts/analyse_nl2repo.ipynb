```python
import jsonlines
```

```python
result_dirs = [
    "/workspace/haoran-cloud/code/kimi-cli/results/GLM-4.7_20260111_143759",
    "/workspace/haoran-cloud/code/kimi-cli/results/GLM-4.7_20260112_134052",
    "/workspace/haoran-cloud/code/kimi-cli/results/GLM-4.7_20260112_170810",
]
```

```python
result_map = {}
for folder in result_dirs:
    with jsonlines.open(f'{folder}/output.eval.jsonl', 'r') as f:
        results = list(f)
    for result in results:
        if not result['instance_id'] in result_map:
            result_map[result['instance_id']] = 0
        result_map[result['instance_id']] = max(result_map[result['instance_id']], result['test_result']['eval_report']['success_rate'])
```
